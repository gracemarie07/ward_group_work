```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.stats import sigma_clip
from pathlib import Path
import re
import warnings
warnings.filterwarnings("ignore", category=UserWarning)


# ----------------------------
# CONFIG
# ----------------------------
input_folder = Path("formirandaandmaddy")  # <-- CHANGE THIS


# ----------------------------
# RA/DEC PARSER (FROM HEADER)
# ----------------------------
def extract_radec(file_path):

    ra, dec = np.nan, np.nan

    with open(file_path, "r") as f:
        for line in f:

            if "Requested input R.A." in line:
                ra = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])

            if "Requested input Dec." in line:
                dec = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])

            if np.isfinite(ra) and np.isfinite(dec):
                break

    return ra, dec


# ----------------------------
# BASELINE FUNCTION
# ----------------------------
def estimate_baseline(flux, sigma=3, maxiters=5):
    flux_clean = flux.replace([np.inf, -np.inf], np.nan).dropna()
    if len(flux_clean) == 0:
        return np.nan
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", message=".*partition.*MaskedArray.*")
        clipped = sigma_clip(flux_clean, sigma=sigma, maxiters=maxiters)
    return np.nanmedian(clipped)

# ----------------------------
# TRANSIENT DETECTION
# ----------------------------
def detect_transient(df_filt):

    det = df_filt["mag_abs"].notna() & np.isfinite(df_filt["mag_abs"])

    if det.sum() < 5:
        return None

    d = df_filt.loc[det].copy()

    idx_peak = d["mag_abs"].idxmin()

    mag_peak = d.loc[idx_peak, "mag_abs"]
    jd_peak = d.loc[idx_peak, "jd"]

    mag_faint = d["mag_abs"].max()

    delta_mag = mag_faint - mag_peak

    return {
        "jd_peak": jd_peak,
        "mag_peak": mag_peak,
        "delta_mag": delta_mag,
        "n_points": len(d)
    }


# ----------------------------
# PROCESS + PLOT ONE FILE
# ----------------------------
def process_and_plot(file_path):

    print(f"\nProcessing {file_path.name}")

    # ---------------- RA/DEC ----------------
    ra_val, dec_val = extract_radec(file_path)

    # ---------------- LOAD ----------------
    df = pd.read_csv(file_path, sep=r"\s+", comment="#", engine="python")
    df.columns = [c.strip().rstrip(",") for c in df.columns]

    df = df.rename(columns={"ccdid": "ccd"})
    if "refFlux" in df.columns:
        df = df.rename(columns={"refFlux": "ref_flux"})

    required = [
        "jd", "forcediffimflux", "forcediffimfluxunc",
        "zpdiff", "filter", "field", "ccd", "qid"
    ]

    for col in required:
        if col not in df.columns:
            raise ValueError(f"{file_path.name}: missing {col}")


    # ---------------- BASELINE ----------------
    df["baseline"] = np.nan
    df["flux_corr"] = np.nan

    group_cols = ["field", "ccd", "qid", "filter"]

    for _, g in df.groupby(group_cols):

        if len(g) < 30:
            continue

        baseline = estimate_baseline(g["forcediffimflux"])

        df.loc[g.index, "baseline"] = baseline
        df.loc[g.index, "flux_corr"] = g["forcediffimflux"] - baseline


    # ---------------- FLUX ----------------
    if "ref_flux" not in df.columns:
        df["ref_flux"] = 0.0

    df["flux_total"] = df["ref_flux"] + df["flux_corr"]

    valid = (
        np.isfinite(df["flux_total"]) &
        np.isfinite(df["forcediffimfluxunc"]) &
        (df["forcediffimfluxunc"] > 0)
    )

    df = df[valid].copy()


    # =========================================================
    # ABSOLUTE PHOTOMETRY
    # =========================================================

    SNT = 3
    SNU = 5

    for col in ["nearestrefmag", "nearestrefmagunc"]:
        if col not in df.columns:
            df[col] = 0.0

    df["nearestrefmag"] = df["nearestrefmag"].fillna(0.0)
    df["nearestrefmagunc"] = df["nearestrefmagunc"].fillna(0.0)

    df["nearestrefflux"] = 10**(0.4 * (df["zpdiff"] - df["nearestrefmag"]))
    df["nearestreffluxunc"] = df["nearestrefmagunc"] * df["nearestrefflux"] / 1.0857

    df["Fluxtot"] = df["forcediffimflux"] + df["nearestrefflux"]

    df["Fluxunctot"] = np.sqrt(
        np.maximum(
            df["forcediffimfluxunc"]**2 - df["nearestreffluxunc"]**2,
            0.0
        )
    )

    bad = df["Fluxunctot"] <= 0

    df.loc[bad, "Fluxunctot"] = np.sqrt(
        df.loc[bad, "forcediffimfluxunc"]**2 +
        df.loc[bad, "nearestreffluxunc"]**2
    )

    df["SNRtot"] = df["Fluxtot"] / df["Fluxunctot"]

    df["mag_abs"] = np.nan
    df["mag_abs_unc"] = np.nan
    df["mag_upperlim"] = np.nan

    det_mask = (df["SNRtot"] > SNT) & (df["Fluxtot"] > 0)
    non_det_mask = ~det_mask

    df.loc[det_mask, "mag_abs"] = (
        df.loc[det_mask, "zpdiff"]
        - 2.5 * np.log10(df.loc[det_mask, "Fluxtot"])
    )

    df.loc[det_mask, "mag_abs_unc"] = 1.0857 / df.loc[det_mask, "SNRtot"]

    df.loc[non_det_mask, "mag_upperlim"] = (
        df.loc[non_det_mask, "zpdiff"]
        - 2.5 * np.log10(SNU * df.loc[non_det_mask, "Fluxunctot"])
    )


    # =========================================================
    # PLOTTING + TRANSIENT FILTER
    # =========================================================

    for filt in df["filter"].dropna().unique():

        df_filt = df[df["filter"] == filt].copy()

        if len(df_filt) == 0:
            continue

        result = detect_transient(df_filt)

        if result is None:
            continue



        # ---------------- PRINT CANDIDATE ----------------
        print("\n==============================")
        print(f"FILTER: {filt}")
        print(f"RA: {ra_val:.6f}, DEC: {dec_val:.6f}")
        print(f"Peak JD: {result['jd_peak']:.3f}")
        print(f"Peak Mag: {result['mag_peak']:.3f}")
        print(f"ΔMag: {result['delta_mag']:.3f}")
        print(f"N points: {result['n_points']}")
        print("==============================\n")


        # ---------------- PLOT ----------------
        plt.figure()

        det = df_filt["mag_abs"].notna()

        plt.errorbar(
            df_filt.loc[det, "jd"],
            df_filt.loc[det, "mag_abs"],
            yerr=np.clip(df_filt.loc[det, "mag_abs_unc"], 0, 0.3),
            fmt="o",
            ms=4,
            label="detections"
        )

        ul = df_filt["mag_upperlim"].notna()

        if ul.any():
            plt.scatter(
                df_filt.loc[ul, "jd"],
                df_filt.loc[ul, "mag_upperlim"],
                marker="v",
                color="red",
                label="upper limits"
            )

        # mark peak
        peak_row = df_filt.loc[df_filt["mag_abs"].idxmin()]
        plt.axvline(peak_row["jd"], color="green", linestyle="--", label="peak")

        plt.title(
            f"{file_path.stem} | {filt}\nRA:{ra_val:.5f} DEC:{dec_val:.5f}"
        )

        plt.gca().invert_yaxis()
        plt.xlabel("JD")
        plt.ylabel("Magnitude")
        plt.legend()

        plt.show()
        plt.close()


# ----------------------------
# RUN BATCH
# ----------------------------
all_files = sorted(input_folder.glob("*.txt"))

for f in all_files:
    try:
        process_and_plot(f)
    except Exception as e:
        print(f"FAILED {f.name}: {e}")
```
