```python
#quering ALeRCE ZTF for AGN objects with probability 0.7 and higher and saving their RA/DEC to a CSV file
from alerce.core import Alerce
import pandas as pd

alerce = Alerce()


page = 1
page_size = 1000   # max allowed per page
all_agn = []

print("Querying ZTF AGN objects...")

while True:
    print(f"Fetching page {page}...")
    
    agn_data = alerce.query_objects(
        classifier="lc_classifier",
        class_name="AGN",
        probability = 0.5,
        page=page,
        page_size=page_size,
        format="pandas"
    )

    if agn_data is None or len(agn_data) == 0:
        break
    
    all_agn.append(agn_data)
    page += 1

# --------------------------------------
# Combine all pages
# --------------------------------------
agn_df = pd.concat(all_agn, ignore_index=True)

print(f"Total AGN retrieved: {len(agn_df)}")

# --------------------------------------
# Keep only relevant columns
# --------------------------------------
agn_coords = agn_df[["oid", "meanra", "meandec", "probability"]]

agn_coords = agn_coords.rename(columns={
    "meanra": "ra",
    "meandec": "dec"
})

# --------------------------------------
# Save to file
# --------------------------------------
output_file = "alerce_agn_ra_dec.csv"
agn_coords.to_csv(output_file, index=False)

print(f"Saved AGN RA/DEC table to {output_file}")
```

```python
# Query LSST objects from ALeRCE and filter AGN probability >= 0.5
from alerce.core import Alerce
import pandas as pd

alerce = Alerce()

classifier_name = "stamp_classifier_rubin_beta"
target_class = "AGN"
prob_threshold = 0.5

page = 1
page_size = 500
all_objects = []

print("Querying LSST objects...")

while True:
    print(f"Fetching page {page}...")

    try:
        obj_page = alerce.query_objects(
            survey="lsst",
            page=page,
            page_size=page_size,
            format="pandas"
        )

    except Exception as e:
        print("Error fetching page:", e)
        break

    if obj_page is None or len(obj_page) == 0:
        print("No more pages.")
        break

    print(f"Page {page} returned {len(obj_page)} objects")

    all_objects.append(obj_page)
    page += 1

# Combine all pages
if len(all_objects) > 0:

    df = pd.concat(all_objects, ignore_index=True)

    # Extract AGN probability column
    prob_col = f"{classifier_name}_{target_class}"

    if prob_col in df.columns:
        agn_df = df[df[prob_col] >= prob_threshold]

        agn_df.to_csv("lsst_agn_rubin.csv", index=False)

        print(f"\nSaved {len(agn_df)} AGN candidates with prob ≥ {prob_threshold}")

    else:
        print(f"Classifier column {prob_col} not found in dataframe.")

else:
    print("\nNo objects returned.")
```
