```python
from astropy.io import fits
import numpy as np

with fits.open("cosmos_catalog_v1.fits") as hdul:
    data = hdul[1].data

    # Clean the class column
    classes = np.char.strip(data["class"].astype(str))

    # Select X-ray AGN
    agn_mask = classes == "X-ray AGN"
    agn_data = data[agn_mask]

    # Extract RA and DEC
    agn_ra = agn_data["ra"]
    agn_dec = agn_data["dec"]

print("Number of X-ray AGN:", len(agn_ra))

# Print first 10 coordinates safely
for i in range(min(10, len(agn_ra))):
    print(f"RA: {agn_ra[i]}, DEC: {agn_dec[i]}")
```

```python
from astropy.io import fits
import pandas as pd

# Open the FITS file
hdul = fits.open("xmmlss.v1.fits")
data = hdul[1].data

# Collect RA/DEC of objects with flag_Xrayagn == 1
ra_list = []
dec_list = []

for obj in data:
    if obj["flag_Xrayagn"] == 1:
        ra_list.append(obj["RA"])
        dec_list.append(obj["DEC"])

hdul.close()

# Create a DataFrame and save to CSV
df = pd.DataFrame({"RA": ra_list, "Dec": dec_list})
df.to_csv("xrayagn_objects.csv", index=False)

print(f"Saved {len(df)} objects to xrayagn_objects.csv")
```
